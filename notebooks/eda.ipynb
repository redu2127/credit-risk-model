```python

# CREDIT RISK MODEL - TASK 2 EDA

# 1. IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

sns.set_style("whitegrid")

print("Libraries loaded successfully")


# 2. LOAD DATASET

df = pd.read_csv("../data/raw/data.csv")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)

df.head()


# 3. DATA OVERVIEW

print("="*50)
print("DATASET INFORMATION")
print("="*50)

df.info()

print("\n")

print("="*50)
print("COLUMN NAMES")
print("="*50)

print(df.columns.tolist())


# 4. FIRST LOOK

display(df.head())

display(df.tail())

display(df.sample(5))


# =========================
# 5. SUMMARY STATISTICS
# =========================

print("="*50)
print("NUMERICAL SUMMARY")
print("="*50)

display(df.describe().T)

print("="*50)
print("CATEGORICAL SUMMARY")
print("="*50)

display(df.describe(include="object").T)


# 6. CHECK MISSING VALUES

missing_count = df.isnull().sum()

missing_percent = (
    df.isnull().sum() / len(df)
) * 100

missing_df = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing Percent": missing_percent
})

missing_df = missing_df.sort_values(
    by="Missing Count",
    ascending=False
)

display(missing_df)

plt.figure(figsize=(10,5))

sns.heatmap(
    df.isnull(),
    cbar=False,
    yticklabels=False
)

plt.title("Missing Values Heatmap")
plt.show()


# 7. DUPLICATES

duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)


# 8. DATA TYPES

datatype_df = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.values
})

display(datatype_df)


# 9. NUMERICAL FEATURES

numerical_cols = df.select_dtypes(
    include=["int64", "float64"]
).columns

print("Numerical Columns")
print(numerical_cols.tolist())


# 10. DISTRIBUTION OF NUMERICAL FEATURES

for col in numerical_cols:

    plt.figure(figsize=(8,4))

    sns.histplot(
        df[col],
        kde=True
    )

    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

    plt.show()


# 11. BOXPLOTS

for col in numerical_cols:

    plt.figure(figsize=(8,4))

    sns.boxplot(
        x=df[col]
    )

    plt.title(f"Boxplot of {col}")

    plt.show()


# 12. CORRELATION ANALYSIS

corr = df[numerical_cols].corr()

plt.figure(figsize=(10,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()


# 13. CATEGORICAL FEATURES

categorical_cols = df.select_dtypes(
    include="object"
).columns

print("Categorical Columns")
print(categorical_cols.tolist())


# 14. CATEGORICAL DISTRIBUTIONS

selected_cols = [
    "ProviderId",
    "ProductCategory",
    "ChannelId",
    "PricingStrategy"
]

for col in selected_cols:

    if col in df.columns:

        plt.figure(figsize=(10,5))

        order = (
            df[col]
            .value_counts()
            .index
        )

        sns.countplot(
            data=df,
            x=col,
            order=order
        )

        plt.title(
            f"Distribution of {col}"
        )

        plt.xticks(rotation=45)

        plt.show()

# 15. TRANSACTION TIME ANALYSIS

df["TransactionStartTime"] = pd.to_datetime(
    df["TransactionStartTime"]
)

df["TransactionHour"] = (
    df["TransactionStartTime"]
    .dt.hour
)

df["TransactionDay"] = (
    df["TransactionStartTime"]
    .dt.day
)

df["TransactionMonth"] = (
    df["TransactionStartTime"]
    .dt.month
)

df["TransactionYear"] = (
    df["TransactionStartTime"]
    .dt.year
)

display(
    df[
        [
            "TransactionHour",
            "TransactionDay",
            "TransactionMonth",
            "TransactionYear"
        ]
    ].head()
)


# 16. TRANSACTIONS BY HOUR

plt.figure(figsize=(12,5))

sns.countplot(
    data=df,
    x="TransactionHour"
)

plt.title(
    "Transactions by Hour"
)

plt.show()


# 17. TRANSACTIONS BY MONTH

plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="TransactionMonth"
)

plt.title(
    "Transactions by Month"
)

plt.show()


# 18. CUSTOMER LEVEL ANALYSIS

customer_summary = (
    df.groupby("CustomerId")
    .agg(
        transaction_count=(
            "TransactionId",
            "count"
        ),
        total_amount=(
            "Amount",
            "sum"
        ),
        avg_amount=(
            "Amount",
            "mean"
        ),
        total_value=(
            "Value",
            "sum"
        ),
        avg_value=(
            "Value",
            "mean"
        )
    )
    .reset_index()
)

display(customer_summary.head())


# 19. CUSTOMER SUMMARY STATS

display(
    customer_summary.describe().T
)


# 20. CUSTOMER TRANSACTION COUNT

plt.figure(figsize=(8,4))

sns.histplot(
    customer_summary["transaction_count"],
    kde=True
)

plt.title(
    "Customer Transaction Count"
)

plt.show()


# 21. TOP CUSTOMERS

top_customers = (
    customer_summary
    .sort_values(
        by="total_value",
        ascending=False
    )
    .head(10)
)

display(top_customers)

# 22. FRAUD ANALYSIS

fraud_counts = (
    df["FraudResult"]
    .value_counts()
)

display(fraud_counts)

fraud_rate = (
    df["FraudResult"]
    .mean()
)

print(
    "Fraud Rate:",
    round(fraud_rate*100,4),
    "%"
)


# 23. FRAUD PLOT

plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="FraudResult"
)

plt.title(
    "FraudResult Distribution"
)

plt.show()


# 24. FRAUD VS AMOUNT

plt.figure(figsize=(8,4))

sns.boxplot(
    data=df,
    x="FraudResult",
    y="Amount"
)

plt.title(
    "Fraud vs Amount"
)

plt.show()


# 25. SAVE CLEAN COPY

df.to_csv(
    "../data/processed/eda_processed.csv",
    index=False
)

print(
    "EDA processed file saved."
)


# FINAL INSIGHTS

print("\n")
print("="*60)
print("TOP 5 EDA INSIGHTS")
print("="*60)

print("""
1. Transaction values are highly skewed.
   Most transactions are small while a few are very large.

2. Customer activity varies significantly.
   Some customers transact frequently while others are inactive.

3. Product categories and channels are imbalanced.
   Certain categories dominate transaction activity.

4. Fraud cases are rare compared with non-fraud cases.
   FraudResult is highly imbalanced.

5. Time-based features such as hour, day, month,
   and year may provide useful behavioral information
   for credit risk modeling.
""")      
```
